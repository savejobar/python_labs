```python
# №1
def ununique_names(filename: str) -> dict:
    """
    Подсчитывает количество повторений каждого слова.

    Args:
        text_of_file (str): Текст для анализа.

    Returns:
        dict: Словарь {слово: количество}.
    """
    if not isinstance(filename, str):
        raise ValueError('file must be a string')
    with open (filename, encoding = 'utf-8') as f:
        text_of_file = f.read()
    
    words = text_of_file.lower().split()
    unique_words = set(words)
    num_words = {}
    for word in unique_words:
        if (len(word) > 2):
            num_words[word.replace(';', '')] = words.count(word)

    return num_words
```

```python
dict_of_unique_names = ununique_names('file_test_of_names.txt')
dict_of_unique_names
```

```python
# №2
def dict_of_words_to_lists_of_tuple (dictionary: dict) -> list:
    list_of_tuples = []
    for word, num_word in dictionary.items():
        list_of_tuples.append((word, num_word))
    
    return list_of_tuples
```

```python
list_of_tuples = dict_of_words_to_lists_of_tuple(dict_of_unique_names)
```

```python
# №3
def list_sort_desc(list_of_tuples: list) -> list:
    list_of_tuples.sort(key = lambda x: x[1], reverse = True)

    return list_of_tuples
```

```python
sortlist = list_sort_desc(list_of_tuples)
```

```python
# №5
def dict_of_repeats(list_of_words: list) -> dict:

    dic_of_words = {}

    for i in range(len(list_of_words) - 1):
        curr = list_of_words[i]
        next = list_of_words[i+1]

        if curr not in dic_of_words:
            dic_of_words[curr] = set()
        dic_of_words[curr].add(next)

    return dic_of_words
```

```python
words_dict = dict_of_repeats(["я", "люблю", "яблоки", "и", "я", "люблю", "груши"])
print(words_dict)
```
